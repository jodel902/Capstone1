```python
plt.plot(df_bit.index,df_bit['Open'])
plt.plot(df_rip.index,df_rip['Open'])
plt.plot(df_eth.index,df_eth['Open'])
plt.xlim([0,1750])
plt.title('Opening Price')
plt.legend(['Bitcoin','Ripple','Ethereum'])
plt.show()
```

```python
plt.plot(df_bit.index,df_bit['High'])
plt.plot(df_rip.index,df_rip['High'])
plt.plot(df_eth.index,df_eth['High'])
plt.xlim([0,1750])
plt.title('High Price')
plt.legend(['Bitcoin','Ripple','Ethereum'])
plt.show()
```

```python
plt.plot(df_bit.index,df_bit['Low'])
plt.plot(df_rip.index,df_rip['Low'])
plt.plot(df_eth.index,df_eth['Low'])
plt.xlim([0,1750])
plt.title('Low Price')
plt.legend(['Bitcoin','Ripple','Ethereum'])
plt.show()
```

```python
plt.plot(df_bit.index,df_bit['Volume'])
plt.plot(df_rip.index,df_rip['Volume'])
plt.plot(df_eth.index,df_eth['Volume'])
plt.xlim([0,200])
plt.y
plt.title('Volume')
plt.legend(['Bitcoin','Ripple','Ethereum'])
plt.show()
```

```python
plt.plot(df_bit.index,df_bit['Market Cap'])
plt.plot(df_rip.index,df_rip['Market Cap'])
plt.plot(df_eth.index,df_eth['Market Cap'])
plt.xlim([0,1750])
plt.ylim([min(df_bit['Market Cap']),max(df_eth['Market Cap'])])
plt.title('Market Cap')
plt.legend(['Bitcoin','Ripple','Ethereum'])
plt.show()
```
